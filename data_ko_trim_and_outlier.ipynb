```python
import pandas as pd                     # outliers bo hote hai jo humre data ke mean ko fluctuate kre
import numpy as np
```

```python
df=pd.read_csv('/content/newplacementdata - newplacementdata.1.csv')
df.head(2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['placement_exam_marks'].describe()
```

```python
sns.boxplot(x=df['placement_exam_marks'])                   # outlier ko detect krne ke liye use m lete hai boxplot and check krne ke liye outliers
```

```python
# finding the ior
percentile25 = df['placement_exam_marks'].quantile(0.25)
percentile75 = df['placement_exam_marks'].quantile(0.75)
percentile25
```

```python
percentile75
```

```python

IQR=percentile75-percentile25
IQR
```

```python
upper_limit=percentile75+1.5*IQR
upper_limit
```

```python
lower_limit=percentile25-1.5*IQR
lower_limit
```

```python
# finding our outliers
df[(df['placement_exam_marks']>upper_limit)]
```

```python
df[(df['placement_exam_marks']<lower_limit)]
```

```python
# trimming -->4(lower)....1 ,upper(25)....(50)----> here 1,50 are outliers-->               outlier ko hannle ke liye trimming and capping
# lower(4) <--> upper(25) mid data.

# capping--> 4 (lower)....1 , upper(25)....(50)----> capping lower point lowest point(1)
# upper point upper point(50) ----> range expand ----> outliers remove.
```

```python
# trimming (outliers removing technique 1)
```

```python
newdf=df[(df['placement_exam_marks']<upper_limit)]
newdf
```

```python
# comparision
```

```python
plt.figure(figsize=(15,7))

plt.subplot(222)
sns.boxplot(x=df['placement_exam_marks'])

plt.subplot(224)
sns.boxplot(x=newdf['placement_exam_marks'])
plt.show()
```

```python
# capping (outlier removing technique 2)
```

```python
new_df_cap=df.copy()
```

```python
# min = 5 , max =15

# min 4,3,1
# max= 20,30,50

# updated_min_value =1
# updated_max_value = 50
```

```python
new_df_cap['placement_exam_marks']=np.where(

new_df_cap['placement_exam_marks']>upper_limit,
upper_limit,

np.where(
    new_df_cap['placement_exam_marks']<lower_limit,
    lower_limit,

new_df_cap['placement_exam_marks']
)
)
```

```python
plt.figure(figsize=(15,8))

plt.subplot(222)
sns.boxplot(x=df['placement_exam_marks'])

plt.subplot(224)
sns.boxplot(x=new_df_cap['placement_exam_marks'])
plt.show()
```
